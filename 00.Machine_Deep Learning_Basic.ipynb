## sigmoid 

- $\displaystyle \sigma(z) = \frac{1}{1+e^{-z}}$

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

array = np.arange(-8, 8, 0.1)
plt.axvline(x=0, color='r', linestyle='--', linewidth=1)        # x=0 y  
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)    # y=0 x  
plt.axhline(y=0.5, color='g', linestyle='--', linewidth=1)      # y=0.5 x  
plt.axhline(y=1.0, color='black', linestyle='--', linewidth=1)  # y=1.0 x  
plt.xlabel('z')          # x label
plt.ylabel('sigmoid(z)') # y label
plt.xticks(np.arange(-8.0, 10.0, 2.0))  # x , start, end, 
plt.yticks(np.arange(0.0, 1.5, 0.5))    # y 
plt.text(-6.0, 0.9, r'$\sigma(z) = \frac{1}{1+e^{-z}}$', fontdict={'size': 18}) #  
plt.plot(array, sigmoid(array))
plt.show()
```

#   (Numerical Derivative) - 1

-    , C/Python  ,     ,   f

$ \displaystyle {f^{'}(x)=\frac{d\overbrace{f(x)}^{\text{ }\\ f(x)\text{ }}}{dx}=\lim_{\Delta{x} \to 0}\frac{f(x+\Delta{x})-f(x)}{\Delta{x}}\frac{df(x)}{dx}=\underbrace{\lim_{\Delta{x} \to 0}}_{\text{   }\\ \Delta x \text{   }} \overbrace{\frac{f(x+\Delta{x})-f(x-\Delta{x})}{2\Delta{x}}}^{\text{/ }}} $

```python
#     
#     
#  x    ,  f     return  
# f:  ,  def lambda  
# x:     ,     
def numerical_derivative(f, x):
  # lim   
  #   0   ,
  # programing 32/64bit  ,
  # 1e-100            .
  # ,  1e-4  0   .
  delta_x = 1e-4
  #     / 
  return (f(x + delta_x) - f(x-delta_x)) / (2 * delta_x)
```

#   (Numerical Derivative) - 2

-       ,    ,       
-   
- 
  - $f(x,y)=2x+3xy+y^{3}$   $x,y$ 2,  ($ \displaystyle \frac{\partial{f}}{\partial{x}}, \frac{\partial{f}}{\partial{y}}$) 
  - $f^{'}(1.0, 2.0)$  
    - $x=1.0$   $y=2.0$  , $\displaystyle \frac{\partial{f(x,2.0)}}{\partial{x}}$ 
    - $y=2.0$   $x=1.0$  , $\displaystyle \frac{\partial{f(1.0,y)}}{\partial{y}}$ 
- $f^{'}(1.0, 2.0) = (8.0, 15.0)$  
  - $x = 1.0$     $y$  2.0   ( ), $x = 1.0$    , $f(x, y)$     
    - , $y = 2.0$    $x = 1.0$   , $f(x, y)$  8.0   
  - $y = 2.0$     $x$  1.0   ( ), $y = 2.0$    , $f(x, y)$     
    - , $x = 1.0$    $y = 2.0$   , $f(x, y)$  15.0   

```python
import numpy as np

#  Machine Learning, TensorFlow, Keras  Deep Learnong Framework    code
# f:  
# x:     numpy (vector or matrix),    
def numerical_derivative(f, x):
  print('debug 1. Initial input variable =', x)

  delta_x = 1e-4
  #     
  #   vector or matrix  
  # 0 
  grad = np.zeros_like(x)
  print('debug 2. Initial grad =', grad)

  #       iterator 
  #  x numpy vector or matrix      
  it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])

  #   
  #  x 2x2 matrix
  # (1,1), (1,2), (2, 1), (2,2)    
  print('=============================================')
  while not it.finished:
    idx = it.multi_index
    print('debug 3. idx =', idx, ', x[idx] =', x[idx])

    #  x numpy type,  mutable type
    #         
    #       .
    tmp_val = x[idx]

    #     
    # idx      .
    # loop       
    #      ...
    x[idx] = float(tmp_val) + delta_x
    print('debug 4. x =', x)
    fx1 = f(x) # f(x + delta_x)

    x[idx] = tmp_val - delta_x
    print('debug 5. x =', x)
    fx2 = f(x) # f(x - delta_x)

    grad[idx] = (fx1 - fx2) / (2 * delta_x)
    print('debug 6. grad[idx] =', grad[idx])
    print('debug 7. grad =', grad)

    #   
    x[idx] = tmp_val
    it.iternext()
    print('=============================================')

  return grad
```
