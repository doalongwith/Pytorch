# 인공지능 머신러닝 딥러닝

- **인공 지능 (Artificial Intelligence)**
    - **인간의 지능을 기계 또는 시스템으로 구현**
    - 인간의 학습능력, 추론능력 등을 컴퓨터를 통해 구현하는 포괄적인 개념
    - '인공지능'이라는 용어 그대로 인간이 지닌 지적 능력을 인공적으로 구현한 것들 모두 인공지능에 포함하여 말한다.
- **머신 러닝 (Machine Learning)**
    - **인공 지능을 구현하는 구체적인 접근 방식**
    - **Regression, Classification, NeuralNetwork, SVM …**
    - 데이터를 이용하여 명시적으로 정의되지 않은 패턴을 학습하여 **미래 결과(값, 분)을 예측**
    - 기계 학습은 기계가 직접 데이터를 학습함으로써 그 속에 숨겨진 일련의 규칙성을 찾는것이다.
    - 머신러닝의 경우 주어진 소재를 우선 인간이 먼저 처리
      - 이미지의 경우 사람이 Training Data(학습 데이터)를 알맞게 분류하는 등 컴퓨터가 인식할 수 있도록 해야 한다. 그 다음 컴퓨터가 데이터에 포함된 특징을 분석하고 축적하는 과정을 거친다.
    - 마지막으로 컴퓨터가 축적된 데이터를 바탕으로 이미지의 특징을 종합해 답을 이끌어 내는 것이 머신러닝이다.
    - Data Mining : 현재 Data간의 상관관계나 속성을 찾는 것이 주목, ML과는 차이가 있다.
    - **Regression(회귀)** - 연속적인 데이터를 학습하여 이 학습된 결과를 바탕으로 미래의 임의의 데이터에 대한 결과를 예측하는 것
    - **Classification(분류)** - Data의 분포를 학습하고 이 학습된 결과를 바탕으로 미래의 임의의 데이터에 대한 분포를 예측하는 것
- **딥러닝 (Deep Learning)**
    - **인공신경망과 빅데이터의 결합**
    - **CNN, RNN, LSTM, …**
    - **Machine Learning의 한 분야로서, 신경망(Neural Network)을 통하여 학습하는 알고리즘의 총집합**
    - 딥러닝은 생물체의 뇌 신호 전달 회로에서 영감을 얻은 **머신러닝 기법 중 하나**이다.
    - 딥러닝의 경우 인간이 하던 작업이 생략된다. 딥러닝은 오로지 데이터로만 배우고, 데이터의 규칙을 찾아내는 수많은 파라미터들을 통해 구현된다. 그렇기 때문에 수많은 데이터가 필요하다. 바탕이 되는 데이터를 그대로 주고, 딥러닝 알고리즘이 회선 신경망(Convolutional Neural Network, CNN)을 이용해 스스로 분석한 후 답을 내는 방식이다.
    - 딥 러닝은 기존 머신러닝과는 다르게 인간 고유의 영역, 이미지 분석, 언어 인식과 같은 직관적이고 고차원적인 사고를 요하는 분야에 강점이 있다.
- **NLP (Natural Language Processing)**
  - 기계가 인간의 언어를 이해하고 처리할 수 있도록 하는 인공지능 분야
- **NLG (Natural Language Generator)**
  - 자연어를 생성하는 데 더 중점을 둔 NLP의 하위 분야
- **LM (Language Model)**
  - 인간의 언어를 이해하고 생성하도록 훈련된 일종의 인공지능 모델
- **LLM (거대 언어 모델, Large Language Model)**
  - 대용량의 언어 모델을 의미
  - 딥 러닝 알고리즘과 통계 모델링을 통해 자연어 처리(Natural Language Processing, NLP) 작업을 수행하는 데에 사용
  - 사전에 대규모의 언어 데이터를 학습하여 문장 구조나 문법, 의미 등을 이해하고 생성할 수 있다.
- ChatGPT
  - OpenAI에서 개발한 LLM 중의 하나, NLU, NLG 작업에 특화


# List Comprehension

```python
x = [ x ** 2 for x in range(10)]
print(x, end='')
```

```python
# if else 조건절이 뒤에 있다
x = [ x for x in range(10) if x % 2 == 0]
print(x, end='')
```

```python
# if else 조건절이 앞에 있고, for 순환문이 뒤에 있음
x = [ x if x % 2 == 0 else 99 for x in range(10)]
print(x, end='')
```

```python
list_2d = [[11, 12],
           [21, 22],
           [31, 32],
           [41, 42]
          ]

# flatten
list_1d = [i for row in list_2d for i in row]
print(list_1d)

# 2D List 를 전치(Transpose) 하기
#list_tranpose = [[row[i] for row in list_2d] for i in range(2)]
list_tranpose = [[row[i] for row in list_2d] for i in range(len(list_2d[0]))]
print(list_tranpose)
```

* 만약 동일하지 않다면, reshape() 이나 transpose(전치행렬) 등을 이용하여 변환 후 사용해야 한다.

### 행렬곱의 사용 예

* 각 반에서 학생 선발시 각 과목에 가중치를 둔다고 하자.
* 가중치를 이용한 3과목의 총합을 구하려면?
* 각 과목별로 (점수 x 가중치)의 총합을 계산하면 되는데,
* 이 경우 행렬곱을 이용하면 편리하다.

---

* 각 학생의 과목별 성적

|성명|국어|수학|영어|
|:---:|---:|---:|---:|
|김|80|90|60|
|이|75|80|90|
|박|90|95|65|
|박|99|70|70|

* 각 반의 과목별 가중치

|과목|우수반|수학반|
|:---:|---:|---:|

```python
score = np.array([[80, 90, 60], [75, 80, 90], [90, 95, 65], [99, 70, 70]])
weight = np.array([[3, 1], [3, 8], [4, 1]])
print('score.shape == ', score.shape)
print('weight.shape == ', weight.shape)
print('score == ', score)
print('weight == ', weight)
```

```python
total = np.dot(score, weight)
print('total.shape == ', total.shape)
print('total == ', total)
```

## 전치행렬(Transpose)

- 원본 matrix의 row는 col으로 col은 row로 바꾸는 것
  - 1행을 1열로, 2행은 2열로

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
B = A.T
print('A.shape ==', A.shape)
print('A ==', A)
print('B.shape ==', B.shape)
print('B ==', B)
```

```python
C = np.array([1, 2, 3, 4, 5]) # vector, not matrix
# C는 vector이므로 transpose가 되지 않음
# vector는 row와 col을 구분할 수 없는 1차원이므로 不可
D = C.T

E = C.reshape(1, 5) # vector를 matrix로 변경
F = E.T

print('C.shape ==', C.shape, 'D.shape ==', D.shape)
print('C.shape ==', E.shape, 'D.shape ==', F.shape)
print('E ==', E)
print('F ==', F)
```

## iterator

- matrix의 모든 원소를 처음부터 끝까지 access하는 경우 사용
- 1차원인 vector에서는 단순한 loop로도 가능하나, 다차원(n-dimension)일 경우 loop를 중복해야 하므로, iterator가 효율적
- MachineLearning에서 손실함수인 Loss Function 구현시에 iterator를 이용해 편미분을 구현


```python
A = np.array([[10, 20, 30, 40], [50, 60, 70, 80]])
print(A, "\n")
print('A.shape ==', A.shape)

# iterator 생성
it = np.nditer(A,  # n-dimensional iterator
               # multi_index: 반복시에 matrix처럼 (roe, col) 형태로 동작, c_inde: c언어 기반의 index
               flags=['multi_index'],
               op_flags=['readwrite']) # 읽고 쓰기 가능, default: readonly
while not it.finished:  # 반복처리가 끝났는지 확인, 끝났으면 True
  idx = it.multi_index  # index를 가져와서
  print('idx:', idx, 'value =>', A[idx])
  it.iternext() # 다음 인덱스로 이동
# 행 단위로 data access
# 2 x 4행렬의 경우
# 먼저 1행부터 access하고 2행 access
# (0, 0) -> (0, 1) -> (0, 2) -> (0, 3) -> (1, 0) -> (1, 1) -> (1, 2) -> (1, 3)
```

## concatenate(...)

- matrix에 row나 col을 추가해 주는 function
- **machine의 Regression(회귀) 구현 시 weight(가중치)나 bias(절편)을 별도로 구현하지 않고 하나의 matrix로 취급하기 위한 프로그래밍 구현 기술**

```python
A = np.array([ [10, 20, 30], [40, 50, 60] ])
print('A.shape ==', A.shape)
print(A)

# A matrix에 row를 추가하려면, 1x3 matrix가 필요
row_add = np.array([70, 80, 90]).reshape(1, 3)
print(row_add.shape)

# A matrix에 col을 추가하려면, 2*1 matix가 필요
col_add = np.array([1000, 2000]).reshape(2, 1)
print(col_add.shape)

# axis = 0 : col 기준
B = np.concatenate((A, row_add), axis=0)
print(B)

# axis = 1 : row 기준
C = np.concatenate((A, col_add), axis=1)
print(C)
```

### sum(), exp(), log()

```python
x = np.array([2, 4, 6, 8])

print('np.sum(x) ==', np.sum(x))
print('np.exp(x) ==', np.exp(x)) # 자연상수 'e'에 대한 지수 함수
print('np.log(x) ==', np.log(x)) # 자연상수 'e'에 대한 로그 함수
```

```python
a = np.array([[1, 0, 3],
             [2, 5, 4],
             [3, 6, 7]])

print(a)
print('np.sum(a) ==', np.sum(a))

# keepdims 인자가 지정되지 않은 경우,
# n-dim array를 input으로 넣으면 (n-1)-dim array가 결과로 반환
col_sum = np.sum(a, axis = 0) # col 기준, col의 sum
print('col_sum ==', col_sum)

row_sum = np.sum(a, axis = 1) # row 기준, row의 sum
print('row_sum ==', row_sum)

# keepdims: n-dim array에서 sum을 구한뒤 그 결과도 n-dim array로 유지
kd_col_sum = np.sum(a, axis = 0, keepdims = True) # col 기준, col의 sum
print('kd_col_sum ==', kd_col_sum)

kd_row_sum = np.sum(a, axis = 1, keepdims = True) # row 기준, row의 sum
print('kd_row_sum ==', kd_row_sum)
```

# 미분(Derivative)

## 의미

- $ \displaystyle {f^{'}(x)=\frac{df(x)}{dx}=\lim_{\Delta{x} \to 0}\frac{f(x)+\Delta{x}}{\Delta{x}}} $

- $\Delta{x}$: **`변화량`**이라는 의미의 그리스문자, Delta라고 읽는다.
- 입력변수 x를 현재 값에서 아주 조금 변화시키면, 함수 f(x)는 얼마나 변화하는가? => 미분계수
- 함수 f(x)는 입력 x의 미세한 변화에 얼마나 민감하게 반응하는가?

image.png

ex) 함수 $f(x) = x^2$ 에서 $f^{'}(x) = 2x$
- $f(3) = 9$ : 입력값 x = 3 이면 출력은 9 임을 의미
- $f^{'}(3) = 6$ : 입력값 x = 3을 미세하게 변화시키면 함수는 현재 임력값의 2배인 6배 변화를 일으키는 것을 의미

## 자주 사용되는 미분 공식

- $f(x)=e^{x} \quad\;\; \Rightarrow f^{'}(x)=e^{x}$
- $f(x)=e^{-x} \quad \Rightarrow f^{'}(x)=-e^{-x}$
- $f(x)=ax^{n} \quad \Rightarrow f^{'}(x)=-nax^{n-1}$
- $f(x)= \ln x \quad \Rightarrow f^{'}(x)=\frac{1}{x}$
  - [예1] $f(x)= 3x^{2} + e^{x} + 7 \quad\; \Rightarrow f^{'}(x)=5x + e^{x} $
  - [예2] $f(x)= \ln x + \frac{1}{x} \quad \quad \quad \Rightarrow f^{'}(x)=\frac{1}{x}-\frac{1}{x^{2}}$ (참고: $\frac{1}{x^{2}}=x^{-2} $) \\
- 참고
  - $f(x)=e^{x}$에서 $e$는 Euler's Number라 불린다. 자연로그(Natural Log)의 밑(base)
  - $e^{x}$: 자연로그($e$)를 밑(base)으로 하는 지수 함수(Exponential Function)
  - 자연 로그((Natural Log)): $\ln{x}, \log_{e}{x}, \log{x}$

# 편미분(Partial Derivative)

## 의미

- 입력변수가 하나 이상인 다변수 함수에서, 미분하고자 하는 변수 하나를 제외한 나머지 변수들은 상수로 취급하고, 해당 변수를 미분하는 것
- 함수 f(x, y)를 변수 x에 대해 편미분하는 경우, 다음과 같이 나타낸다.

  - $ \displaystyle  \frac{\partial{f(x, \;y)}}{\partial{x}} $

- 입력변수는 x, y 2개가 있는데, 이 중 y는 상수 취급하고, x에 대해서만 함수를 미분하겠다는 의미
- $\partial$ : 편미분 기호

## 사용예

- $\displaystyle  f(x,y)=2x+3xy+y^{3}$
  - [예1] 함수 $f(x,y)$를 $x$에 대해 편미분하면?
    - $ \displaystyle  \frac{\partial{f(x,y)}}{\partial{x}} = \frac{(\partial{2x+3xy+y^{3})}}{\partial{x}}=2+3y$
  - [예2] 함수 $f(x,y)$를 $y$에 대해 편미분하면?
    - $ \displaystyle  \frac{\partial{f(x,y)}}{\partial{x}} = \frac{(\partial{2x+3xy+y^{3})}}{\partial{x}}=3x+3y^{2}$

 - [예3] 체중함수가 `체중(야식, 운동)`처럼 야식과 운동에 영향을 받는 2개의 변수를 가지는 함수라 가할 경우, 편미분을 사용하면 각 변수의 변화량에 따른 체중 변화량을 구할 수 있다.

# Chain Rule(연쇄 법칙)

- 합성함수를 미분하기 위해, **합성함수를 구성하는 각 함수의 미분의 곱**으로 나타내는 방식


## - 합성 함수(composite function)

- 여러 함수로 구성된 함수
  - $ \displaystyle { f(x)=e^{3x^{2}} \quad \Rightarrow e^t, \; t=3x^2 }$
  - $ \displaystyle { f(x)=e^{-x} \quad \Rightarrow e^t, \; t=-x } $

## Chain Rule의 예

- `Chain Rule`: 합성함수를 미분할 경우, 분모와 분자에 동일한 변수값을 곱해주고, 각각 미분해서, 최종적으로 각각의 미분값을 곱해주는 형태를 의미

- $\displaystyle f(x)=e^{3x^{2}}$를 chain rule로 미분하는 경우, $t=3x^2$ 으로 놓으면, $\displaystyle f(x)=e^{t}$
  - $ \displaystyle \frac{\partial{f}}{\partial{x}}=\overbrace{\frac{\partial{f}}{\partial{t}}\frac{\partial{t}}{\partial{x}}}^{\text{chain rule 적용} \\ \text{(약분 개념)}}=\frac{\partial{(e^t)}}{\partial{t}}\underbrace{\frac{\partial{(3x^2)}}{\partial{x}}}_{t=3x^2 \text{ 대입}}=(e^t)(6x)=(\underbrace{e^{3x^{2}}}_{t=3x^2 \text{ 대입}})(6x)=6xe^{3x^{2}} $
- $f(x)=e^{-x}$를 chain rule로 미분하는 경우, $t=-x$ 으로 놓으면, $f(x)=e^{t}$
  - $ \displaystyle \frac{\partial{f}}{\partial{x}}=\overbrace{\frac{\partial{f}}{\partial{t}}\frac{\partial{t}}{\partial{x}}}^{\text{chain rule 적용} \\ \text{(약분 개념)}}=\frac{\partial{(e^t)}}{\partial{t}}\underbrace{\frac{\partial{(-x)}}{\partial{x}}}_{t=-x \text{ 대입}}=(e^t)(-1)=(\underbrace{e^{-x}}_{t=-x \text{ 대입}})(-1)=-e^{-x}$

# 수치 미분 (Numerical Derivative) - 1

- 수학 공식을 쓰지 않고, C/Python 등을 이용하여, 주어진 값이 미세하게 변할 경우, 함수 값 f가 얼마나 변하는지를 계산하는 것을 의미
- 우선, 입력변수 하나인 간단한 함수 구현

$ \displaystyle {f^{'}(x)=\frac{d\overbrace{f(x)}^{\text{미분하려는 함수}\\ f(x)\text{ 정의}}}{dx}=\lim_{\Delta{x} \to 0}\frac{f(x+\Delta{x})-f(x)}{\Delta{x}}\frac{df(x)}{dx}=\underbrace{\lim_{\Delta{x} \to 0}}_{\text{극한 개념을 구하기 위해}\\ \Delta x \text{는 작은 값으로 설정}} \overbrace{\frac{f(x+\Delta{x})-f(x-\Delta{x})}{2\Delta{x}}}^{\text{분모/분자 구현}}} $

```python
# 수치미분을 계산해 주는 함수 정의
# 입력 변수가 하나인 수치미분 함수
# 입력값 x가 아주 미세하게 변화할 경우, 함수 f는 얼마나 변화하는지 수치적으로 계산해서 return해 주는 함수
# f: 미분하려는 함수, 외부에서 def나 lambda 등으로 정의
# x: 미분값을 알고자 하는 입력 값, 즉 미세하게 변화하는 입력 값
def numerical_derivative(f, x):
  # lim에 해당하는 작은 값
  # 이론적으로는 거의 0인 값을 입력해야 하나,
  # programing에서는 32/64bit로 제한되어 있고,
  # 1e-100 같은 너무 작은 갓을 사용하면 실제 계산에서 반올림 오차가 발생할 수 있다.
  # 그래서, 경험적으로 1e-4 정도면 0에 수렴한다고 알려져 있다.
  delta_x = 1e-4
  # 미분 정의 나타난 수식대로 분자/분모를 구현
  return (f(x + delta_x) - f(x-delta_x)) / (2 * delta_x)
```

# 수치 미분 (Numerical Derivative) - 2

- 입력변수가 하나 이상인 다 변수 함수의 경우, 입력변수는 서로 독립적이기 때문에, 수치미분 또한 변수의 갯수만큼 개별적으로 계산해야 함
- 실무에서 일반적으로 사용됨
- 예제
  - $f(x,y)=2x+3xy+y^{3}$ 에서 입력변수 $x,y$ 2개이므로, 각각 수치미분($ \displaystyle \frac{\partial{f}}{\partial{x}}, \frac{\partial{f}}{\partial{y}}$)을 수행
  - $f^{'}(1.0, 2.0)$을 계산하기 위해서는
    - $x=1.0$에서의 미분계수는 변수 $y=2.0$을 상수로 대입하여, $\displaystyle \frac{\partial{f(x,2.0)}}{\partial{x}}$을 수행
    - $y=2.0$에서의 미분계수는 변수 $x=1.0$을 상수로 대입하여, $\displaystyle \frac{\partial{f(1.0,y)}}{\partial{y}}$을 수행
- $f^{'}(1.0, 2.0) = (8.0, 15.0)$ 의 의미
  - $x = 1.0$ 에서의 미분값을 구한다는 것은 $y$ 값은 2.0 으로 고정한 상태(상수로 유지)에서, $x = 1.0$ 을 미세하게 변화시킬 경우, $f(x, y)$ 가 얼마나 변화하는지 알아보겠다는 의미
    - 즉, $y = 2.0$ 으로 고정된 상태에서 $x = 1.0$ 을 미세하게 변화시키면, $f(x, y)$ 는 8.0 만큼 변화한다는 의미
  - $y = 2.0$ 에서의 미분값을 구한다는 것은 $x$ 값은 1.0 으로 고정한 상태(상수로 유지)에서, $y = 2.0$ 을 미세하게 변화시킬 경우, $f(x, y)$ 가 얼마나 변화하는지 알아보겠다는 의미
    - 즉, $x = 1.0$ 으로 고정된 상태에서 $y = 2.0$ 을 미세하게 변화시키면, $f(x, y)$ 는 15.0 만큼 변화한다는 의미

```python
import numpy as np

# 다양한 Machine Learning, TensorFlow, Keras 등의 Deep Learnong Framework 에서 실제 사용되는 code
# f: 다변수 함수
# x: 모든 변수를 포함하고 있는 numpy 객체(vector or matrix), 하나 이상의 값을 가짐
def numerical_derivative(f, x):
  print('debug 1. Initial input variable =', x)

  delta_x = 1e-4
  # 계산된 수치미분 값을 저장할 변수
  # 입력으로 들어온 vector or matrix와 동일한 크기
  # 0으로 초기화
  grad = np.zeros_like(x)
  print('debug 2. Initial grad =', grad)

  # 모든 변수에 대해 편미분 하기 위해 iterator 생성
  # 입력값 x가 numpy의 vector or matrix이므로 처음부터 끝싸지 모든 값을 가리키기 위해
  it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])

  # 입력변수의 갯수만큼 반복
  # 입력변수 x가 2x2 matrix라면
  # (1,1), (1,2), (2, 1), (2,2) 의 순으로 편미분을 계산
  print('=============================================')
  while not it.finished:
    idx = it.multi_index
    print('debug 3. idx =', idx, ', x[idx] =', x[idx])

    # 변수 x가 numpy type이고, 따라서 mutable type이다
    # 따라서 함수내에서 값이 변경될 수 있어 원래 값을 보관
    # 아래에서 미분값을 계산하는 과정에서 원래 값이 변경된다.
    tmp_val = x[idx]

    # 하나의 변수에 대한 수치미분 계산
    # idx에 해당하는 순서의 변수에 대한 수치미분을 구한다.
    # loop의 첫번째에서는 첫번째 변수만 증감하고 나머지는 그대로 전달
    # 두번째에서는 두번째 변수만 증감하여 전달 ...
    x[idx] = float(tmp_val) + delta_x
    print('debug 4. x =', x)
    fx1 = f(x) # f(x + delta_x)

    x[idx] = tmp_val - delta_x
    print('debug 5. x =', x)
    fx2 = f(x) # f(x - delta_x)

    grad[idx] = (fx1 - fx2) / (2 * delta_x)
    print('debug 6. grad[idx] =', grad[idx])
    print('debug 7. grad =', grad)

    # 보관했던 값을 복원
    x[idx] = tmp_val
    it.iternext()
    print('=============================================')

  return grad
```

## 예제 1

변수가 하나인 함수 $f(x)=x^{2}$에서 $f^{'}(3.0)$ 는?

```python
# 입력변수가 1개인 함수 f(x) = x ** 2 정의
def fun1(input_obj):
  # input_obj - vector or matrix인 numpy object
  x = input_obj[0]
  return x ** 2

result = numerical_derivative(fun1, np.array([3.0]))
print(result)
```

## 예제 3

변수가 4개인 함수 $f(w,x,y,z)=wx+xyz+3w+xy^{2}$에서 $f^{'}(1.0, 2.0, 3.0, 4.0)$ 은?

```python
# 입력변수가 4개인 함수 f(w, x, y, z) = wx + xyz + 3w + zy^2 정의
# input_obj는 matrix
def func3(input_obj):
  w = input_obj[0, 0]
  x = input_obj[0, 1]
  y = input_obj[1, 0]
  z = input_obj[1, 1]

  return ( w * x + x * y * z + 3 * w + z * np.power(y, 2) )

# 입력을 2x2 matric으로 구성
input = np.array([[1.0, 2.0], [3.0, 4.0]])
result = numerical_derivative(func3, input)
```
