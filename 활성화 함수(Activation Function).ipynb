# 1.이진 활성화 함수(Binary step activation function)

- threshold-based function: 0을 기준으로 output을 나눔
- 다중 출력이 불가능

  $h(x)=
\begin{cases}
-x,\;if\;x<0\\
+x,\;if\;x\geq0
\end{cases}$

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_step_activate_function(x):
    y = x > 0
    return y.astype(int)

x = np.arange(-2, 2, 0.1)
y = binary_step_activate_function(x)

plt.figure(figsize=(3, 3))
plt.plot(x,y)
plt.show()
```

# 2.선형 활성화 함수 (Linear activation function)

- backpropagation(활성화 함수를 미분하는 과정을 통해 손실 값을 줄이는 과정) 사용 불가
    - 선형 함수의 미분값은 항상 같은 상수값이 나와 사용 불가능
- 활설화 함수 본래의 역활인 입력을 normaliztion하여 간단하게 하는 것을 수행하지 못함

  $h(x)=ax$

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_step_activate_function(x):
    return x

x = np.arange(-2, 2, 0.1)
y = binary_step_activate_function(x)

plt.figure(figsize=(3, 3))
plt.plot(x,y)
plt.show()
```

# 3.비선형 활성화 함수 (Non-linear activation function)

- backpropagation 사용 가능
- 활성화 함수의 normalization 수행, 다중 출력이 가능

## 3.1.시그모이드 활성화 함수(Sigmoid activation function)

- 장점
  - 유연한 미분 값
    - 입력에 따른 급격한 값의 변화가 없다.
  - 출력 값의 범위가 0과 1 사이로 제한됨으로써 정규화 중 exploding gradient 문제 방지    
    - 미분식이 단순한 형태 지님 / 깔끔한 예측 값
      - 계산 값이 양 끝단으로 가까울수록 0과 1에 수렴
- 단점
  - vanishing gradient 문제 발생
    - 입력이 아무리 커도 미분 값의 범위가 0과 1/4 사이로 제한됨으로써 층이 많을수록 gradient 값이 0에 수렴
  - 출력의 중심이 0이 아님
    - 뉴런으로 들어오는 데이터가 항상 양수인 경우 gradient는 모두 양수 또는 음수가 된다.
    - 이는 gradient 업데이트 중 지그재그로 변동하는 문제점이 발생,  이 지그재그로 변동하는 gradient는 학습 효율성을 감소 시킬 수 있다.

  $h(x)=\frac{1}{1+e^{-x}}$

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_step_activate_function(x):
    return 1/(1+np.exp(-x))

x = np.arange(-10, 10, 0.1)
y = binary_step_activate_function(x)

plt.figure(figsize=(3, 3))
plt.plot(x,y)
plt.show()
```

## 3.2.하이퍼볼릭 탄젠트(Hyperbolic tangent/Tang)

- 특징은 Sigmoid 함수와 유사
- 다만, 출력의 중심이 0에 있다는 장점이 존재
- 이는 시그모이드 함수의 단점 중 출력의 중심이 0이 아니라 발생하는 문제점 보완

  $h(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_step_activate_function(x):
    return np.tanh(x)

x = np.arange(-10, 10, 0.1)
y = binary_step_activate_function(x)

plt.figure(figsize=(3, 3))
plt.plot(x,y)
plt.show()
```

## 3.3.렐루 활성화 함수 (ReLU/Rectified Linear Unit activation function)


  $h(x)=\begin{cases}0,\;if\;x<=0\\x,\;if\;x>0\end{cases}$

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_step_activate_function(x):
    return np.maximum(0, x)

x = np.arange(-10, 10, 0.1)
y = binary_step_activate_function(x)

plt.figure(figsize=(3, 3))
plt.plot(x,y)
plt.show()
```
