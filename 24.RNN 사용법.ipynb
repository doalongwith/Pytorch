```python
import torch

# batch_size=1
input_data = torch.Tensor([ [ [1], [2], [3] ] ]) # (batch_size=1, time_steps=3, featrue=1)
RnnModel = torch.nn.RNN(
              input_size=1,  # 입력: 1, 1개의 feature
              hidden_size=3, # 출력: 3, 3개의 hidden state, 따라서 출력도 3개
              batch_first=True)

# 은닉층의 수(num_layers): default=1
outputs, last_hs = RnnModel(input_data)
# outputs: RNN의 모든 시점(timestamp)에서의 hidden state, [h1, h2, h3]
# outputs은 (batch_size, time_stpes, hidden_size) = (1, 3, 3)
# last_hs: 마지막 시점(timestamp)의 hidden state, [h3]

print(f'input_data.size() = {input_data.size()}') # input_data.size() = torch.Size([1, 3, 1]), (batch_size=1, time_steps=3, featrue=1)
print(f'outputs.size() = {outputs.size()}')       # outputs.size() = torch.Size([1, 3, 3]), (batch_size=1, time_stpes=3, hidden_size=3)
print(f'outputs = {outputs}')
print(f'last_hs.size() = {last_hs.size()}')       # last_hs.size() = torch.Size([1, 1, 3]), (num_layers=1, batch_size=1, hidden_size=3)
print(f'last_hs = {last_hs}')
```
